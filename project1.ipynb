```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("Global YouTube Statistics.csv", encoding = "latin1")
```

```python
pd.set_option("display.float_format", "{:.2f}".format)
```

```python
category_subscribers_mean = df.groupby("category")["subscribers"].mean()
category_subscribers_mean
```

```python
category_subscribers_mean.sort_values(ascending = False, inplace = True)
```

# Data Visulisation

```python
category_subscribers_mean.head(7).plot(kind = "bar")
plt.title("Mean Subscribers by Category")
plt.xlabel("Category")
plt.ylabel("Mean Subscribers")
plt.xticks(rotation = 90)
plt.show()
```

```python
channel_type_counts = df["channel_type"].value_counts()
channel_type_counts
```

```python
channel_type_counts.head().plot(kind = "pie", autopct= "%1.1f%%")
plt.title("Channel Type Distrubition")
plt.show()
```

```python
plt.scatter(df["subscribers"], df["video views"])
plt.title('Scatter Plot of Subscribers vs Video Views')
plt.xlabel("Subscribers")
plt.ylabel("Video Views")
plt.show()
```

```python
country_subscribers_total = df.groupby("Country")["subscribers"].sum().sort_values(ascending = False)
```

```python
country_subscribers_total.head(10).plot(kind = "bar")
plt.title("Total Sibscribers by Country")
plt.xlabel("country")
plt.ylabel("Total Subscribers")
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.boxplot(x= "channel_type", y= "highest_monthly_earnings", data= df)
plt.title("Distribution of Highest Monthly Earnings by Channel Type")
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.scatter(df["Gross tertiary education enrollment (%)"], df["subscribers"], alpha = 0.5)
plt.title("Scatter plot of subcribers vs gros tertiary education enrollment")
plt.xlabel("Gross tertiary education enrollment")
plt.ylabel("Subscribers")
plt.show()
```

```python
df.groupby("channel_type")["Unemployment rate"].mean().sort_values(ascending= False).plot(kind = "bar")
plt.title("Average Unemployment Rate by Channel Type")
plt.xlabel("Channel Type")
plt.ylabel("Average Unompleyment Rate")
plt.xticks(rotation = 90)
plt.show()
```

```python
df.groupby(["created_year", "channel_type"])["subscribers"].mean().unstack().tail().plot(kind = "bar", width = 1)
plt.title("Average Subscribers and Created Year by Channel Type")
plt.xlabel("Created Year")
plt.ylabel("Average Subscribers")
plt.xticks(rotation = 45)
plt.legend(title= "Channel Type", bbox_to_anchor = (1.0, 1), loc = "upper left")
plt.show()
```

```python
df.groupby(["created_year", "created_month"])["subscribers"].sum().unstack().head().plot(kind = 'area', stacked = True)
plt.title("Subscribers Growth Over Time")
plt.xlabel("Year")
plt.ylabel("Total Subscribers")
plt.legend(title = "Month", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
df.groupby(["created_year", "Country"])["highest_yearly_earnings"].mean().unstack().head().plot(kind = "line", marker = "o")
plt.title("Trend of Highest Yearly Earnings by Created Year and Country")
plt.xlabel("Created Year")
plt.ylabel("Average Highest Yearly Earnings")
plt.legend(title = "Country", bbox_to_anchor = (1.05, 1), loc = "upper left")
plt.xticks(rotation = 45)
plt.show()
```

```python
category_highest_monthly_earnings_sum = df.groupby("category")["highest_monthly_earnings"].sum()
```

```python
category_highest_monthly_earnings_sum.sort_values(ascending= False, inplace= True)
```

```python
category_highest_monthly_earnings_sum.head().plot(kind = "pie", autopct = "%1.1f%%")
plt.title("Percentage of Highest Monthly Earnings by Category")
plt.ylabel("")
plt.show()
```

```python
channel_created_subscribers_mean = df.groupby(["channel_type", "created_year"])["subscribers_for_last_30_days"].mean().unstack()
```

```python
channel_created_subscribers_mean
```

```python
sns.heatmap(channel_created_subscribers_mean.head(5), cmap = "YlGnBu", annot = True)
plt.title("Mean Subscribers for Last 30 Days by Channel Type and Created Year")
plt.xlabel("Created Year")
plt.ylabel("Channel Type")
plt.show()
```

```python
df.groupby("created_month")["subscribers_for_last_30_days"].mean().plot(kind = "line", marker = "o")
plt.title("Average Subscribers for the Last 30 Days by Month")
plt.xlabel("Month")
plt.ylabel("Average Subsribers for the Last 30 Days")
plt.xticks(rotation = 45)
plt.show()
```
